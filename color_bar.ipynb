```python
import seaborn as sns
import pandas as pd

colors = sns.color_palette("viridis", n_colors=60)
colors_hex = sns.color_palette(colors).as_hex()
dates_month = pd.date_range(start='2020-01', end='2024-12', freq='MS') # Dates

year = 2020
print(f'\t{year}\n')
for i, color in enumerate(colors_hex):
    print(f"{dates_month[i].strftime('%Y-%m')}: {color}")
    if (i + 1) % 12 == 0 and year != 2024:
        year += 1
        print(f'\n\t{year}\n')
```
